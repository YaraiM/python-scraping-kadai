```python
import requests
from bs4 import BeautifulSoup

# urlを定義
url = 'https://news.yahoo.co.jp/articles/2568a5cdde645e4ab2e265368b7d35ae555a4370'

# urlからHTMLオブジェクトのtext属性を取得
response = requests.get(url)

# Responseオブジェクトのtext属性を解析
soup = BeautifulSoup(response.text, 'html.parser')

# 本文の要素を取得
# デベロッパツールのコピー機能でセレクタをコピーできる
body_element = soup.select_one('#uamods > div.article_body.highLightSearchTarget > div > p')

# 本文のテキストを取得
body_text = body_element.text

# 本文を表示
print(body_text)
```
